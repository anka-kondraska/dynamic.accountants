```python
from xbrl import XBRLParser, GAAP, GAAPSerializer, DEISerializer
```

```python
xbrl_parser = XBRLParser()
xbrl = xbrl_parser.parse('google10k2019.xml')
```

```python
gaap_obj = xbrl_parser.parseGAAP(xbrl, doc_date="20190930", ignore_errors=0)
```

```python
serializer = GAAPSerializer()
result = serializer.dump(gaap_obj)
```

```python
print(result)
```
