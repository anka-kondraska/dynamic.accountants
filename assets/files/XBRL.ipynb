```python
all_tables=soup.findAll()
```

```python
fin_df_columns = ["Name", "Context", "Text", "Id", "Unitref", "Decimals"]
```

```python
 fin_df = pd.DataFrame(columns=fin_df_columns)
```

```python

for tag in all_tables:
    if tag.name.startswith('dei') or tag.name.endswith('TextBlock'):
        continue
    if tag.get('contextref'):
        lt=tag.get('contextref').split('_')
        print(tag.name,lt)
        row = {
            fin_df_columns[0]: tag.name, 
            fin_df_columns[1]: lt, 
            fin_df_columns[2]: tag.text, 
            fin_df_columns[3]: tag.get('id'),
            fin_df_columns[4]: tag.get('unitref'),
            fin_df_columns[5]: tag.get('decimals'),

        }
        fin_df = fin_df.append(row, ignore_index=True)
```

```python
from xbrl import XBRLParser, GAAP, GAAPSerializer, DEISerializer
```

```python
xbrl_parser = XBRLParser()
xbrl = xbrl_parser.parse(open('google10k2019.xml'))
```

```python
gaap_obj = xbrl_parser.parseGAAP(xbrl, doc_date="20190930", context="current", ignore_errors=0)
```

```python
serializer = GAAPSerializer()
result = serializer.dump(gaap_obj)
```

```python
print(result)
```

```python
dei_obj = xbrl_parser.parseDEI(xbrl)
serializer = DEISerializer()
result1 = serializer.dump(dei_obj)
print(result1)
```
